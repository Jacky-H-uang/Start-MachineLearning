```python
import numpy as np
import pandas as pd
train = pd.read_csv('./datasets/train.csv')
display(train.head())
```

```python
# look submission data files
submission = pd.read_csv('./datasets/gender_submission.csv')
display(submission.head())
```

```python
# read the test data files
test = pd.read_csv('./datasets/test.csv')
display(test.head())
```

```python
women = train.loc[train.Sex == 'female']['Survived']

# 活下来的女人的 rate
rate_women = sum(women) / len(women)
rate_women
```

```python
men = train.loc[train.Sex == 'male']['Survived']

# 活下来的男人的 rate
rate_men = sum(men) / len(men)
rate_men
```

```python
"""
分离数据将训练数据的 80% 来训练算法 20% 来验证算法
"""
def split_train_test(data,test_ratio = 0.8):
    shuffled_indices = np.random.permutation(len(data))     # random 的产生 len(data) 的 array
    test_set_size = int(len(data) * test_ratio)
    train_indices = shuffled_indices[test_set_size:]
    test_indices = shuffled_indices[:test_set_size]

    return  data.iloc[train_indices] , data.iloc[test_indices]  # 分别返回训练集合和测试集合
train_data , test_data = split_train_test(train)
traget = test_data.pop('Survived')
display(test_data)
```

```python
"""
随机森林的算法
"""

from sklearn.ensemble import RandomForestClassifier

features = ["Pclass", "Sex", "SibSp", "Parch"]
X_train = pd.get_dummies(train_data[features])
y_train = train_data['Survived']
X_test = pd.get_dummies(test_data[features])

rnd_clf = RandomForestClassifier(n_estimators = 100 , max_depth = 5 , random_state = 42)
rnd_clf.fit(X_train , y_train)
rnd_clf.predict(X_test)
```

```python
"""
使用交叉验证测试精度
"""
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
from sklearn.base import clone
from sklearn.model_selection import StratifiedKFold

# StratifielKFold 执行分成抽样
skfolds = StratifiedKFold(n_splits = 3 , random_state = 42 , shuffle = True)

y_train_pred = cross_val_score(rnd_clf , X_train , y_train , cv = 3)
y_train_pred
```

```python
# Test the model

model = RandomForestClassifier(n_estimators = 100, max_depth = 5, random_state = 42)
X = pd.get_dummies(train[features])
y = train['Survived']
X_test = pd.get_dummies(test[features])
model.fit(X,y)
predictions = model.predict(X_test)
```

```python
# Save and Commit the Model

output = pd.DataFrame({'PassengerId': test.PassengerId , 'Survived': predictions})
output.to_csv('my_submission.csv', index = False)
print("Your submission was successfully saved!")
```
