```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
from cv2 import cv2
```

```python
base_path = './Stanford_DogSets/Images'
labels = set()
paths = []
for d in os.listdir(base_path):
    labels.add(d)
    paths.append((base_path + '/' + d, d))
for l in labels:
    print(l)
```

```python
# 加载预处理图片
def load_preprocess_images(path):
    image = cv2.imread(path)
    image = cv2.resize(image, (224, 224))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

```python
# 准备好训练的 X , y
X, y = [], []
i = 0
for path, label in paths:
    i += 1
    if i == 18 or i == 23 or i == 41 or i == 49 or i == 90:    continue 
    if path == "./Images/.DS_Store":     continue
    for image_path in os.listdir(path):
        image = load_preprocess_images(path + "/" + image_path)
        X.append(image)
        y.append(label)
```

```python
# clean the label's name of y
Y = []
for name in y:
    Y.append(name.split('-')[1])
Y
```

```python
# 标签二进制化
encoder = LabelBinarizer()
y = encoder.fit_transform(np.array(y))
y
```

```python
# split 训练集 80% 用来训练 20% 用来测试
from sklearn.model_selection import train_test_split
X = np.array(X)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 42)
```
